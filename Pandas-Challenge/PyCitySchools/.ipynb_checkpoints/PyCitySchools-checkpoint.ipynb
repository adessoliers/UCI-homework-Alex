```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python

school_data_complete.head()
```

```python
# Convert to integers
school_data["budget"] = pd.to_numeric(school_data["budget"])
school_data["size"] = pd.to_numeric(school_data["size"])
school_data_complete["reading_score"] = pd.to_numeric(school_data_complete["reading_score"])
school_data_complete["math_score"] = pd.to_numeric(school_data_complete["math_score"])
school_data_complete["size"] = pd.to_numeric(school_data_complete["size"])
school_data_complete["budget"] = pd.to_numeric(school_data_complete["budget"])

# Total # of schools
total_number_schools = len(school_data_complete["school_name"].unique())

# Total # students
total_number_students = len(school_data_complete["Student ID"].unique())

#Total Budget
total_budget_unique = school_data_complete["budget"].unique()
total_budget = total_budget_unique.sum()

# Average math score
avg_math_score = school_data_complete["math_score"].mean()

# Average reading score
avg_reading_score = school_data_complete["reading_score"].mean()

# Percent of passing math score >= 70
percent_math_pass = (school_data_complete.loc[school_data_complete["math_score"]>=70]
["math_score"].count()/total_number_students)*100

# Percent of passing reading csore >= 70
percent_reading_pass = (school_data_complete.loc[school_data_complete["reading_score"]>=70]
                        ["reading_score"].count()/total_number_students)*100

# Overall passing scores
overall_pass = (percent_reading_pass+percent_math_pass)/2

# DATAFRAME
district_summary = pd.DataFrame({"Total Schools": [total_number_schools],
                                "Total Students": [total_number_students],
                                "Total Budget": [total_budget],
                                "Average Math Score": [avg_math_score],
                                "Average Reading Score": [avg_reading_score],
                                "% Passing Math": [percent_math_pass],
                                "% Passing Reading": [percent_reading_pass],
                                "% Overall Passing": [overall_pass]})


# Formatting
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

district_summary
```

```python
#Variables
school_name = school_data_complete.set_index('school_name').groupby(['school_name'])
school_type = school_data.set_index('school_name')['type']
total_studs = school_name['Student ID'].count()
total_budg = school_data.set_index('school_name')['budget']
budget_per_stu = total_budg/total_studs
avg_math_score2 = school_name["math_score"].mean()
avg_reading_score2 = school_name["reading_score"].mean()
percent_math_pass2 = school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/total_studs
percent_reading_pass2 = school_data_complete.loc[school_data_complete["reading_score"]>=70.].groupby("school_name")["Student ID"].count()/total_studs
overall_pass2 = (percent_math_pass2+percent_reading_pass2)/2

school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_studs,
    "Total School Budget": total_budg,
    "Per Student Budget": budget_per_stu,
    "Average Math Score": avg_math_score2,
    "Average Reading Score": avg_reading_score2,
    "% Passing Math": percent_math_pass2,
    "% Passing Reading": percent_reading_pass2,
    "% Overall Passing": overall_pass2
})

# Formatting
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("{:,}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("{:.2f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.1f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.1f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:.2%}".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.2%}".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map("{:.2%}".format)

school_summary.head()
```

```python
school_summary["Average Math Score"] = pd.to_numeric(school_summary["Average Math Score"])
school_summary["Average Reading Score"] =  pd.to_numeric(school_summary["Average Reading Score"])
school_summary["Per Student Budget"] =  pd.to_numeric(school_summary["Per Student Budget"])
school_summary["% Passing Math"] =  school_summary["% Passing Math"].replace('%', '', regex=True).astype('float')
school_summary["% Passing Math"] =  pd.to_numeric(school_summary["% Passing Math"])
school_summary["% Passing Reading"] =  school_summary["% Passing Reading"].replace('%', '', regex=True).astype('float')
school_summary["% Passing Reading"] =  pd.to_numeric(school_summary["% Passing Reading"])
```

```python
# Creating bins
bins = [0,585,630,645,675]
spend_range = ["< $584", "$585-629", "$630-644", "$645-675"]

#pd.cut
school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"],
                                                        bins=bins, labels=spend_range)

# Group by spend range
group_spend_range = school_summary.groupby(["Spending Ranges (Per Student)"])

# Calculations
mean_math = group_spend_range["Average Math Score"].mean()
mean_reading = group_spend_range["Average Reading Score"].mean()
mean_percent_math = group_spend_range["% Passing Math"].mean()
mean_percent_reading = group_spend_range["% Passing Reading"].mean()
overall_percent_pass = (mean_percent_math+mean_percent_reading)/2

score_by_school_spending = pd.DataFrame({"Average Math Score":mean_math,
                                          "Average Reading Score":mean_reading,
                                          "% Passing Math":mean_percent_math,
                                          "% Passing Reading":mean_percent_reading,
                                          "% Overall Passing":overall_percent_pass})

# Re-Format
score_by_school_spending["Average Math Score"] = score_by_school_spending["Average Math Score"].map("{:,.2f}".format)
score_by_school_spending["Average Reading Score"] = score_by_school_spending["Average Reading Score"].map("{:,.2f}".format)
score_by_school_spending["% Passing Math"] = score_by_school_spending["% Passing Math"].map("{:.2f}".format)
score_by_school_spending["% Passing Reading"] = score_by_school_spending["% Passing Reading"].map("{:.2f}".format)
score_by_school_spending["% Overall Passing"] = score_by_school_spending["% Overall Passing"].map("{:.2f}".format)

score_by_school_spending

```

```python
### SCORE SBY SCHOOL SIZES ###
```

```python
bins1 = [0,1000,2000,5000]
size_range = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#Data into categories
school_summary["Total Students"] = school_summary["Total Students"].replace(',', '', regex=True).astype('float')
school_summary['Total Students'] = pd.to_numeric(school_summary["Total Students"])
school_summary["School Size"] = pd.cut(school_summary["Total Students"], bins = bins1, labels=size_range)

# Group data by school size
by_school_size = school_summary.groupby(["School Size"])

# Calculate passing rates by school size
mean_math_score = by_school_size["Average Math Score"].mean()
mean_reading_score = by_school_size["Average Reading Score"].mean()
mean_percent_math = by_school_size["% Passing Math"].mean()
mean_percent_reading = by_school_size["% Passing Reading"].mean()
overall_percent_pass2 = (mean_math_score + mean_reading_score)/2

# DataFrame
by_school_size_df = pd.DataFrame({"Average Math Score": mean_math_score.map("{:,.2f}".format),
                                  "Average Reading Score": mean_reading_score.map("{:,.2f}".format),
                                  "% Passing Math": mean_percent_math.map("{:.2f}".format),
                                  "% Passing Reading": mean_percent_reading.map("{:.2f}".format),
                                  "% Overall Passing": overall_percent_pass2.map("{:.2f}".format)
                                 })

by_school_size_df


```
